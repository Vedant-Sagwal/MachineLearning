```python
df = pd.read_csv("./dataCSV/Online Sales Data.csv")
df = df.rename(columns={'Date': 'NewDate'})
df.drop("Unit Price", axis=1)
df['New Unit Price'] = df['Unit Price'].apply(lambda x : x * 2)
df.head()
```

```python
grouped_mean = df.groupby('Product Name')['Total Revenue'].mean()
print(grouped_mean)
```

```python
grouped_sum = df.groupby(['Product Name', 'Product Category'])['Total Revenue'].sum()
print(grouped_sum)
```
