```python
import matplotlib.pyplot as plt
```

```python
x = [1,4,9,16,25]
y = [1,2,3,4,5]
plt.plot(x,y)
plt.xlabel("squares")
plt.ylabel("square roots")
plt.title("Sqaures and Square roots")
plt.show()
```

```python
x = [1,4,9,16,25]
y = [1,2,3,4,5]
plt.plot(x,y,color="red", marker="o")
```

```python
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,8,27,64,125]

plt.subplot(2,2,1)
plt.plot(x, y1, color="orange")
plt.subplot(2,2,2)
plt.plot(y1, y2, color="red")
plt.subplot(2,2,3)
plt.plot(y1, y2, color="green")
```
